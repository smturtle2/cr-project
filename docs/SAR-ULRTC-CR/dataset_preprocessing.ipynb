## SEN12MS-CR Dataset Preprocessing

1. Download the SEN12MS-CR dataset from [this link](https://mediatum.ub.tum.de/1554803).
2. Unzip all the *.tar.gz files. The unzipped folders should look like this:
    ```
    SEN12MS-CR/
          |- ROIs1158_spring_s1/
          |- ROIs1158_spring_s2_cloudy/
          |- ROIs1158_spring_s2/
          ...
    ```
3. Run the code blocks below to preprocess the dataset and obtain the train/val/test splits in .npy format.
4. Download additional translated image files from [this link](https://223.195.240.67:1006/sharing/vGqxpjjlZ) (Only accessed using campus network) and paste the folders into the same directory as the preprocessed data.

The data folder after finishing preprocessing should look like this:

    SEN12MS-CR_1/
            |- SEN12MSCR_train_input_npy/
            |- SEN12MSCR_train_gt_npy/
            |- SEN12MSCR_train_sar_npy/
            |- SEN12MSCR_train_mask_npy/
            |- SEN12MSCR_train_translated/
            |- SEN12MSCR_train_input_npy_full/
            |- SEN12MSCR_train_gt_npy_full/
            |- SEN12MSCR_train_sar_npy_full/
            |- SEN12MSCR_train_mask_npy_full/
            |- SEN12MSCR_train_translated_full/
          ...

### 10000 training samples split

````python
import torch
from dataLoader import SEN12MSCR_2
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import torch
from dataLoader import SEN12MSCR
import numpy as np
from tqdm import tqdm
import os
from torch.utils.data import Subset, DataLoader
import numpy as np
import matplotlib.pyplot as plt
import cv2
import random
import torch
import scipy.io

train_path = 'SEN12MS-CR'

sen12mscr_train       = SEN12MSCR(train_path, split='train', region='all')

seed = 42  # Set the seed value
torch.manual_seed(seed)  # Set the seed for torch
np.random.seed(seed)  # Set the seed for numpy

subset_indices = torch.randperm(len(sen12mscr_train))[:10000]

sen12mscr_test_subset = Subset(sen12mscr_train, subset_indices)
data_test_loader      = torch.utils.data.DataLoader(sen12mscr_test_subset, batch_size=1, shuffle=True, num_workers=4)

output_dir = 'SEN12MS-CR_1/SEN12MSCR_train_input_npy'
out_dir_gt = 'SEN12MS-CR_1/SEN12MS-CR/SEN12MSCR_train_gt_npy'
out_dir_sar = 'SEN12MS-CR_1/SEN12MS-CR/SEN12MSCR_train_sar_npy'
out_dir_mask = 'SEN12MS-CR_1/SEN12MS-CR/SEN12MSCR_train_mask_npy'

os.makedirs(output_dir, exist_ok=True)
os.makedirs(out_dir_gt, exist_ok=True)
os.makedirs(out_dir_sar, exist_ok=True)
os.makedirs(out_dir_mask, exist_ok=True)

# Clear data in the directories
for directory in [output_dir, out_dir_gt, out_dir_sar, out_dir_mask]:
    for file in os.listdir(directory):
        file_path = os.path.join(directory, file)
        if os.path.isfile(file_path):
            os.remove(file_path)

for pdx, (samples) in enumerate(tqdm(data_test_loader)):

    file_dir = samples['input']['S2 path'][0]
    filename = os.path.basename(file_dir)

    np.save(os.path.join(output_dir, filename).replace('.tif', '.npy'), samples['input']['S2'][0].cpu().numpy())
    np.save(os.path.join(out_dir_gt, filename).replace('.tif', '.npy'), samples['target']['S2'][0].cpu().numpy())
    np.save(os.path.join(out_dir_sar, filename).replace('.tif', '.npy'), samples['input']['S1'][0].cpu().numpy())
    np.save(os.path.join(out_dir_mask, filename).replace('.tif', '.npy'), samples['input']['masks'][0].cpu().numpy())
````
